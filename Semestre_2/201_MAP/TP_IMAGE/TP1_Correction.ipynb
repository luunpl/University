## Image créée par programmation

Une image peut aussi être créée pixel par pixel en utilisant une boucle d’instructions.

**Exemple**:

```python
# Création d’une image de dimension 50 x 50 avec tous les pixels noirs
image = np.zeros([50,50]) # (zero = noir)
# Mettre en blanc certains pixels
for i in range(50):
    for j in range(50):
        # pixel blanc si i+j multiple de 7
        if np.mod(i+j,7)==0:
            image[i,j] = 255
# affichage de l’image
show_image(image)
```

• Exemple : de la même manière, créez une image de taille 100 × 100 d’un disque noir de centre (xC = 50, yC = 50) et rayon r = 40 sur fond blanc :

<p style="text-align: center;">un pixel(i,j) est noir si et seulement si `(i−yC)2 +(j−xC)2 <r2 </p>


```python
# Création d’une image de dimension 100 x 100 avec tous les pixels blancs
# Dimensions de l’image
L = 100; H = 100
image = np.ones([H,L]) * 255

# Paramètres du disque
xC = 50
yC = 50
r = 40

for i in range(H):
    for j in range(L):
        #pixel noir s’il appartient au disque
        if (i-yC) ** 2 + (j-xC) ** 2 < r ** 2:
            image[i,j] = 0

show_image(image)
```

## Correction d'une image en niveaux de gris.
On peut transformer une image en modifiant la valeur de chaque pixel par une même fonction d’une variable réelle $p \mapsto (p)$.
**Exemple** : pour avoir le négatif d’une image, la fonction est $f(p) = 255 − p$. Dans la plupart des cas, les fonctions standard de Python ou Numpy peuvent s’appliquer directement à tous les éléments d’un tableau, sans qu’il soit nécessaire de les parcourir un par un.

**Remarque** : les images au format bitmap ne doivent en principe comporter que des valeurs entières entre 0 et 255. Néanmoins lors des calculs des valeurs non entières peuvent apparaître, et la fonction `afficher_image` permet quand même d’afficher les images correspondantes. Par contre, les valeurs doivent rester dans l’intervalle [0, 255]. Ci-dessous, un exemple de fonction ainsi qu’une procédure qui effectue la conversion en nombres compris entre 0 et 255.

```python
# Fonction ramenant les valeurs entre 0 et 255
def min0_max255(x):
    return np.clip(x, 0, 255)

# définition de la fonction f
def f(p):
    return 255-p

# Tracé de la fonction
t = np.arange(256)
plt.figure() # creer une nouvelle fenêtre graphique
plt.plot(t,f(t)) #  tracer la fonction en rouge

# lecture et affichage de l'image 
image = plt.imread("bastille.bmp")
show_image(image)

# Opération sur image
image2 = f(image)     # appliquer la fonction
image2 =  min0_max255(image2) # repasser à des valeurs entre 0 et 255 
# affichage de l'image modifiée
show_image(image2)
```

```python
# Exercice 4
### A COMPLETER - DEBUT ###
def modifier_image(image,p0=0,p1=255):
    N,M = image.shape
    new_image = image.copy()
    for i in range(N):
        for j in range(M):
            val = image[i,j]
            if val <= p0 :
                new_image[i,j] = 0
            elif p0 < val and val <= p1:
                # Eq. de la droite
                new_image[i,j] = (val - p0) * 255 / (p1 - p0)
            else:
                new_image[i,j] = 255
    return new_image

### A COMPLETER - FIN ###

show_image(image)
show_image(modifier_image(image,0,150))

```

La correction dite Gamma s’applique aux valeurs des pixels normalisées (ramenées à l’intervalle $[0,1]$ en divisant par $255$), par l’intermédiaire de la fonction

$$ h:\left\lbrace \begin{array}{rcl} [0,1] & \rightarrow & [0,1]\\v & \mapsto & h(v)=v^\gamma \end{array} \right.$$

Le cas $\gamma = 1$ correspond à la fonction identité. La fonction $f$ à appliquer aux valeurs de pixels est donc

$$f (p) = 255 × h(p/255).$$

## Opérations entre 2 images
On peut effectuer une opération entre 2 images par une fonction $f$ sur les niveaux de gris, c’est-à-dire les valeurs de pixels normalisées comme pour la correction Gamma (ramenées à l’intervalle $[0,1]$ en divisant par $255$) :
$$v' =f(v_1,v_2)$$
Par la suite, on utilisera les opérations entre images suivantes :

| Opération | Notation | Fonction correspondante |
| :-: | :-: | --- |
| produit | $F(p_1,p_2) = p_1 \otimes p_2$ | $f(v_1,v_2)=v_1 \times v_2$ |
| somme | $F(p_1,p_2) = p_1 \oplus p_2$ | $f(v_1,v_2)=\min(v_1 + v_2,1)$ |

L'opération _négatif_, notée $p'=N(p)$ correspond à la fonction $v'=f(v) = 1 - v$.

La cellule ci-dessous effectue le produit des images `papillon.bmp` et `degrade-1.bmp`. 

```python
# lecture des images 
papillon = plt.imread("papillon.bmp")
degrade1 = plt.imread("degrade-1.bmp")
degrade2 = plt.imread("degrade-2.bmp")
canal_1 = plt.imread("canal_1.bmp")
canal_2 = plt.imread("canal_2.bmp")
canal_3 = plt.imread("canal_3.bmp")
desert = plt.imread("desert.bmp")
ciel = plt.imread("ciel.bmp")
masque = plt.imread("masque-desert.bmp")

fig,ax = plt.subplots()
ax.imshow(255 * np.clip(papillon / 255 * degrade1 / 255,0., 1.),cmap='gray')
ax.set_axis_off()
```

## Exercice 7
Modifier la cellule suivante pour calculer les images suivantes:
| `papillon.bmp` $\otimes$ `degrade-2.bmp`| produit des deux images. |
| --- | --- |
| `papillon.bmp` $\oplus$ `degrade-1.bmp`| somme des deux images. |
| `papillon.bmp` $\oplus$ `degrade-2.bmp`| somme des deux images. |
| `canal_1.bmp` $\oplus$ `canal_2.bmp`$\oplus$ `canal_3.bmp`| somme des trois images. |
| `desert.bmp` $\otimes$ `masque-desert.bmp`| produit des deux images. |
| `desert.bmp` $\otimes$ N(`masque-desert.bmp`)| produit des deux images. |

```python
# Exercice 7

### A COMPLETER - DEBUT ###
# produit entre IMAGE1 et IMAGE2
def produit(im1,im2):
    v1 = im1 / 255
    v2 = im2 / 255
    vm = np.clip(v1 * v2,0,1)
    return 255 * vm

def somme(im1,im2):
    v1 = im1 / 255
    v2 = im2 / 255
    vm = np.clip(v1 + v2,0,1)
    return 255 * vm

def negation(im):
    return 255 - im

# affichage de l'image modifiée
fig,ax = plt.subplots(1,5,figsize=(15,3))
ax[0].imshow(papillon,cmap='gray');ax[0].set_axis_off();ax[0].set_title('Image 1')
ax[1].imshow(degrade1,cmap='gray');ax[1].set_axis_off();ax[1].set_title('Image 2')
ax[2].imshow(produit(papillon,degrade1),cmap='gray');ax[2].set_axis_off();ax[2].set_title("IMAGE1 x IMAGE2")
ax[3].imshow(somme(papillon,degrade1),cmap='gray');ax[3].set_axis_off();ax[3].set_title("papillon + degrade1")
ax[4].imshow(negation(papillon),cmap='gray');ax[4].set_axis_off();ax[4].set_title("Negation de IMAGE1")

### A COMPLETER - FIN ###
```
