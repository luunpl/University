```python
import numpy as np #importe les outils de gestion des tableaux
import matplotlib.pyplot as plt #importe les outils d'affichages matplotlib
plt.rcParams['figure.dpi'] = 120 #reglage de resolution des figures
plt.rcParams['savefig.dpi'] = 240 #reglage de resolution des figures enregistrees

N=10#declare une variable N initialisee sur la valeur 10
X=np.arange(N) #cree le tableau [0,1,2,3,4,5,....,N-1]. X[i]=i pour i=0,1,...,N-1
Y=np.random.random(N) #cree un tableau de N nombre aleatoires entre 0 et 1, numerotes de 0 a N-1
Z=np.linspace(1,3,N) #cree un tableau de N nombres qui subdivisisent l'intervalle [1,3] avec un ecart regulier

plt.plot(X,Y) #Trace le graphe de Y en fonction de X, en rejoignant les points (X[i],Y[i]) pour i=0,..,N-1. Par defaut le graphe est represente en tracant les segments [(X[i],Y[i]),(X[i+1],Y[i+1])] pour i=0,1,..,N-2, cela peut etre change en explorant la documentation
plt.plot(X,Z,label='graphe de Z') #Trace le graphe de Z en fonction de X, avec en plus une legende associee a ce graphe.
plt.title("Affichage de Y,Z en fonction de X") #Donne un titre a la figure
plt.legend(loc='upper center') #localisation de l'affichage des legendes
plt.xlabel("abscisses (X)") #Nomme l'axe des abscisses
plt.ylabel("ordonnees (Y,Z)") #Nomme l'axe des ordonnees
plt.show() #affichage de la figure
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams['figure.dpi'] = 120 
plt.rcParams['savefig.dpi'] = 240 

def h(x,y):#definition d'une fonction de deux variables
    return (-np.sin(2*y+x),np.sqrt(1+x**2+y**4))#expression de h(x,y)
    #Rappel de notations de fonctions classiques:
    #x*y:multiplication
    #x**y: x exposant y
    #np.cos, np.sin, np.tan, np.log, np.exp, np.sqrt, np.abs: quelques fonctions classiques

def Affichage_Champs(g, xmin,xmax,ymin,ymax,N):
    #g: fonction qui definit le champs de vecteurs
    #xmin,xmax: intervalle d'abscisses affichees
    #ymin,ymax: intervalle d'ordonnees affichees
    #N: nombre de pas de discretisation
    x=np.zeros([N+1,N+1])
    y=np.zeros([N+1,N+1])
    u=np.zeros([N+1,N+1])
    v=np.zeros([N+1,N+1])#initialisation de quatres tableaux indexes sur [0,N]x[0,N]
    for i in range(N+1):#boucle for qui parcourt i=0,1,2,...,N
        for j in range(N+1):#idem pour j=0,1,2,..,N
            x[i,j],y[i,j]=xmin+(xmax-xmin)*i/N,ymin+(ymax-ymin)*j/N#abscisse et ordonnees du point associe a (i,j)
            u[i,j],v[i,j]=g(x[i,j],y[i,j])#valeur du champs de vecteur en (x[i,j],y[i,j])
    u=u*(ymax-ymin)/(xmax-xmin)#Changement d'echelle.
    u,v=u/np.sqrt(0.01+u**2+v**2),v/np.sqrt(0.01+u**2+v**2)#Normalisation du champs de vecteur: cela permet de rendre plus visible la direction.
    plt.quiver(x,y,u,v)#affichage le champs de vecteur: cette fonction trace le vecteur (u[i,j],v[i,j]) base en (x[i,j],y[i,j]), pour i,j qui parcourent [0,N]x[0,N]


Affichage_Champs(h,-2,2,-2,2,20)#Appelle la fonction ci-dessus
```
